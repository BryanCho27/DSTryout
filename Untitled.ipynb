```python
#EDA(Exploratory Data Analysis)
#For getting summary of our dataset
#To find some pattern from our dataset
import matplotlib.pyplot as plt

def bar_graph(data):
    survived = data[data["Survived"]==1]
    dead = data[data["Survived"]==0]
    sm = len(survived[survived["Sex"]=="male"])
    sf = len(survived[survived["Sex"]=="female"])
    dm = len(dead[dead["Sex"]=="male"])
    df = len(dead[dead["Sex"]=="female"])
    bar_width = 0.35
    index=range(2)
    plt.bar([0,1],[dm,df],width=bar_width, label="Dead")
    plt.bar([0+bar_width,1+bar_width],[sm,sf],width=bar_width, label="Survived")
    plt.legend()
    plt.xticks([0+bar_width/2,1+bar_width/2],["Male","Female"])
    plt.show()
    
bar_graph(data)
```

```python
def bar_graph2(data):
    survived = data[data["Survived"]==1]
    dead = data[data["Survived"]==0]
    s1 = len(survived[survived["Pclass"]==1])
    s2 = len(survived[survived["Pclass"]==2])
    s3 = len(survived[survived["Pclass"]==3])
    d1 = len(dead[dead["Pclass"]==1])
    d2 = len(dead[dead["Pclass"]==2])
    d3 = len(dead[dead["Pclass"]==3])
    bar_width = 0.35
    plt.bar([0,1,2],[d1,d2,d3],width=bar_width, label="Dead")
    plt.bar([0+bar_width,1+bar_width,2+bar_width],[s1,s2,s3],width=bar_width, label="Survived")
    plt.legend()
    plt.xticks([0+bar_width/2,1+bar_width/2,2+bar_width/2],["Class 1","Class 2","Class 3"])
    plt.show()

bar_graph2(data)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
def preprocessing(text):
    text=text.lower()
    #0. remove punct
    pattern = r'[^\w\s]'
    nopunct = re.sub(pattern,"",text)
    #1. tokenization
    tokens=word_tokenize(nopunct)
    #2. remove stopwords
    stopword_lst = set(stopwords.words("english"))
    filtword = [word for word in tokens if word not in stopword_lst]
    #3. Stemming
    #stemmer = PorterStemmer()
    #stemmed_words = [stemmer.stem(word) for word in filtword]
    
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtword]
    result = " ".join(lemmatized_words)
    return result

print(preprocessing("The green frog jumps over the lazy dog."))
```

```python
s = "This is awesome"
result = preprocessing(s)
lst = result.split()
positive = 0
negative = 0
for word in lst:
    synsets = wn.synsets(word)
    synset = synsets[0]
    sentiment = sn.senti_synset(synset.name())
    if sentiment:
        positive += sentiment.pos_score()
        negative += sentiment.neg_score()
print(positive/len(lst))
print(negative/len(lst))
```

```python
# Countvectorizer - convert the text into vector based on the word frequency

text = ["Hello my name is scott, this is my python code","scott is a student","features of count vectorizer"]

from sklearn.feature_extraction.text import CountVectorizer 
import pandas as pd
count_vect = CountVectorizer()
matrix = count_vect.fit_transform(text)
arr = matrix.toarray()
df = pd.DataFrame(data=arr,columns = count_vect.get_feature_names_out())
df
```

```python
#Sport news vs Econ news
#tf-idf vectorizer = Term Frequency-Inverse Document Frequency
#Countvectorizer --> most frequent word = most important
#Tf-idf --> statistical method that evaluates importance of each word in a document relative to the entire data

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
count_vect = TfidfVectorizer()
matrix = count_vect.fit_transform(text)
arr = matrix.toarray()
df = pd.DataFrame(data=arr,columns = count_vect.get_feature_names_out())
df
```

```python
import pandas as pd
data = pd.read_csv("IMDB Dataset.csv")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
def preprocessing(text):
    text=text.lower() 
    pattern = r'[^\w\s]'
    nopunct = re.sub(pattern,"",text)
    tokens=word_tokenize(nopunct)
    stopword_lst = set(stopwords.words("english"))
    filtword = [word for word in tokens if word not in stopword_lst]
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in filtword]
    result = " ".join(stemmed_words)
    return result
```

```python
data["preprocess"] = data["review"].apply(preprocessing)
data.to_csv("IMDB_processed.csv")
data
```
